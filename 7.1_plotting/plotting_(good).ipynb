```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.linspace(-np.pi,np.pi,100)
cos_x=np.cos(x)
sin_x=np.sin(x)
plt.plot(x,cos_x,label='cos x')
plt.plot(x,sin_x,label='sin x')
plt.legend()
```

```python
fig,ax=plt.subplots(nrows=1,nrows=2)
ax[0].plot(x,cos_x,label='cos x',c='r',ls='--')
```
