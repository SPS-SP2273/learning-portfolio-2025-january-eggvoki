# Task 4

```python
import numpy as np
import pandas as pd

# Function to shorten names
def clean(text):
    try:
        text = text.replace("National University of Singapore", "NUS")
        text = text.replace("Nanyang Technological University", "NTU")
        text = text.replace("Singapore Management University", "SMU")
        text = text.replace("Singapore Institute of Technology", "SIT")
        text = text.replace("Singapore University of Technology and Design", "SUTD")
        text = text.replace("Singapore University of Social Sciences", "SUSS")
        return text
    except AttributeError:
        return text  # Handles NaN (which are floats)

# Apply the cleaning function only to the university column (adjust column name)
df['university'] = df['university'].apply(clean)

# Replace 'na' with np.nan across all columns
df.replace('na', np.nan, inplace=True)

# Convert basic_monthly_median to float
df['basic_monthly_median'] = df['basic_monthly_median'].astype(float)

df
```
