```python
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Step 1: Load the WAV file
def load_wav(filename):
    rate, data = wav.read(filename)
    return rate, data

# Step 2: Perform FFT to convert time-domain to frequency-domain
def perform_fft(data, rate):
    fft_data = np.fft.fft(data)
    freqs = np.fft.fftfreq(len(fft_data), 1 / rate)
    
    # Only keep the positive frequencies (real part)
    positive_freqs = freqs[:len(freqs) // 2]
    positive_fft_data = np.abs(fft_data[:len(fft_data) // 2])
    
    return positive_freqs, positive_fft_data

# Step 3: Find the peaks (identify the prominent frequencies)
def identify_peaks(freqs, fft_data, threshold=0.1):
    peaks, _ = find_peaks(fft_data, height=threshold)
    peak_freqs = freqs[peaks]
    peak_amplitudes = fft_data[peaks]
    return peak_freqs, peak_amplitudes

# Step 4: Main code to load and analyze one WAV file
def analyze_wav(filename):
    rate, data = load_wav(filename)
    print(f"Loaded {filename} with sample rate {rate} Hz")

    # If the audio is stereo, take just one channel (mono)
    if len(data.shape) > 1:
        data = data[:, 0]

    # Perform FFT
    freqs, fft_data = perform_fft(data, rate)

    # Identify peaks in the FFT data (main frequencies)
    peak_freqs, peak_amplitudes = identify_peaks(freqs, fft_data)

    # Plot the frequency spectrum
    plt.figure(figsize=(10, 6))
    plt.plot(freqs, fft_data, label="FFT")
    plt.scatter(peak_freqs, peak_amplitudes, color='r', label="Peaks")
    plt.title(f"Fourier Transform of {filename}")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.xlim(0, 5000)  # Limit x-axis for clarity, adjust as needed
    plt.show()

    # Output the frequencies of the peaks
    print(f"Identified peak frequencies: {peak_freqs}")
    print(f"Peak amplitudes: {peak_amplitudes}")

# Example usage: Replace 'your_file.wav' with the path to your actual file
analyze_wav("Violin_quaver_240_beat.wav")
```
